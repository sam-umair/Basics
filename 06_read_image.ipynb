```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
plt.plot()
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y= [12,34,17,40,32,50,22]
plt.plot(x,y)
plt.show()
```

### Method 1

```python
fig = plt.figure()
axis = fig.add_subplot()
axis.plot(x,y)
plt.show()
```

### Method 2

```python
fig = plt.figure()
axis = fig.add_axes([1,1,1,1])
axis.plot(x,y)
plt.show()
```

### Method 3

```python
fig,axis = plt.subplots()
axis.plot(x,y)
plt.show()
```

```python
# Prepare the data
x = [1,2,3,4,5,6,7]
y= [12,34,17,40,32,50,22]
#set a plot environment
home,room = plt.subplots()
#plot the graph
room.plot(x,y)
#Customization
room.set(title = 'A Good Plot',
         xlabel = 'x-axis',
         ylabel = 'y-axis'
         )
plt.show()
```

### Plot multiple graphs

```python

home,(room1,room2) = plt.subplots(nrows=1,ncols=2,figsize=(10,5))

```

```python

home,(room1,room2) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

```

```python

home,((room1,room2),(room3,room4)) = plt.subplots(nrows=2,ncols=2,figsize=(10,5))

room1.plot(x,y)
room4.plot(x,y)
```
